전처리 과정

```python
df['No-show'].replace("No", 0,inplace=True)
df['No-show'].replace("Yes", 1,inplace=True)
```

```python
# df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
# df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])

# # Create a variable called "AwaitingTime" by subtracting the date the patient made the appointment and the date of the appointment.
# df['Reservation_wait_days'] = df["AppointmentDay"].sub(df["ScheduledDay"], axis=0)
# df["Reservation_wait_days"] = (df["Reservation_wait_days"] / np.timedelta64(1, 'D')).abs()
```

```python
df['Reservation_wait_days'] = (pd.to_datetime(df['AppointmentDay']).dt.date - pd.to_datetime(df['ScheduledDay']).dt.date).dt.days
```

```python
df['No-show'].info()
```

```python
df["Reservation_wait_days"].describe()
```

음수 값 확인 => 절대값으로 처리

```python
df["Reservation_wait_days"] = df["Reservation_wait_days"].abs()
```

```python
df["Reservation_wait_days"].describe()
```

```python
df['Total_cancellations'] = df.groupby('PatientId')['No-show'].apply(lambda x: x.cumsum())
```

```python
df['Handcap'] = pd.Categorical(df['Handcap'])
Handicap = pd.get_dummies(df['Handcap'], prefix = 'Handicap')
df = pd.concat([df, Handicap], axis=1)
```

```python
df
```

```python
# 'gender' 열을 원핫 인코딩하여 새로운 열로 추가합니다.
gender_encoded = pd.get_dummies(df['Gender'], prefix='Gender')

# 기존의 'gender' 열을 제거합니다.
df.drop('Gender', axis=1, inplace=True)

# 원핫 인코딩된 열을 원래의 데이터프레임에 추가합니다.
df = pd.concat([df, gender_encoded], axis=1)
```

```python
df = df[(df.Age >= 0) & (df.Age <= 100)]
```

```python
df
```

```python
df.drop(["PatientId",'AppointmentID',
         'ScheduledDay','Handcap','AppointmentDay','Neighbourhood'], axis=1, inplace = True)
```
